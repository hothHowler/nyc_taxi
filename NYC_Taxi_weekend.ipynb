## Comparison of Weekday to Weekend Ridership (6am-9pm)

```python
import pandas as pd 
import glob 
import random
import numpy as np
import progressbar
import fiona 
import shapely
from shapely.geometry import Point, shape
%matplotlib inline
import seaborn as sns
from nyc_taxi import *
```

```python
fils = glob.glob('./data/*loc.pk')
```

```python
fils[1]
dat = dfilts(fils[1],dowmax=5, dowmin=1, hourmax=21, hourmin=6)
opdwd=plotHeatPickDrop(dat,tit=' Weekdays 6am-9pm',nrm=5)
opdwd.to_pickle('2014-05-Weekdays.pk')
```

```python
np.mean(opdwd.values),np.sum(opdwd.values)
```

```python
dat = dfilts(fils[1],dowmax=7, dowmin=7, hourmax=21, hourmin=6)
sunday= plotHeatPickDrop(dat,tit=' Sunday 6am-9pm',nrm=1)
sunday.to_pickle('2014-05-sunday.pk')
```

```python
np.mean(sunday.values), np.sum(sunday.values)
```

```python
dat = dfilts(fils[1],dowmax=6, dowmin=6, hourmax=21, hourmin=6)
saturday = plotHeatPickDrop(dat,tit=' Saturday 6am-9pm',nrm=1)
saturday.to_pickle('2014-05-saturday.pk')
np.mean(saturday.values),np.sum(saturday.values)
```

```python
saturday.mean()
```

```python
sns.heatmap(((saturday)-(opdwd)), linewidths=0.1,vmin=-500,vmax=500)
```

## Compare Late Night Weekends to Weekday 

```python
dat = dfilts(fils[1],dowmax=6, dowmin=5, hourmax=24, hourmin=21)
opdwehln=plotHeatPickDrop(dat,tit=' Weekend 9pm-12am',nrm=3*3)
opdwehln.to_pickle('2014-05-Weekend-latenight.pk')
```

```python
dat = dfilts(fils[1],dowmax=4, dowmin=1, hourmax=24, hourmin=21)
opdwdhln=plotHeatPickDrop(dat,tit=' Weekend 9pm-12am',nrm=3*3)
opdwdhln.to_pickle('2014-05-Weekday-latenight.pk')
```

```python
diff_wewdln = (opdwehln)-(opdwdhln)
diff_wewdln.columns.name='Pickup Neighborhood (Weekend-Weekday 9pm-12am)'
sns.heatmap(diff_wewdln, linewidths=0.1,vmin=-40,vmax=40)
```
