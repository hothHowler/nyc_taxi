## Comparison of Early Evening to Late Night ridership

```python
import pandas as pd 
import glob 
import random
import numpy as np
import progressbar
import fiona 
import shapely
from shapely.geometry import Point, shape
%matplotlib inline
import seaborn as sns
from nyc_taxi import *
```

```python
fils = glob.glob('./data/*loc.pk')
```

```python
fils[1]
dat = dfilts(fils[1],dowmax=5, dowmin=1, hourmax=24, hourmin=21)
ndow = dow_counter(dat)
opdwdhln=plotHeatPickDrop(dat,tit='',nrm=np.sum(ndow[0:4])*3)
opdwdhln.to_pickle('2014-05-Weekdays-latenight.pk')
```

```python
fils[1]
dat = dfilts(fils[1],dowmax=5, dowmin=1, hourmax=21,hourmin=18)
opdwdhev=plotHeatPickDrop(dat,tit=' Weekdays 6pm-9pm',nrm=np.sum(ndow[0:4])*3)
opdwdhev.to_pickle('2014-05-Weekdays-evening.pk')
```
